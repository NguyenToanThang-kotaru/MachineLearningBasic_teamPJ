```python
from os.path import join
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..")))
from model.catboost import catboost
from model.xgboost import xgboost

data_dir = join('..', '..', 'data')
input_dir = join(data_dir, 'preprocessed')
output_dir = join(data_dir, 'feature_engineered')

train = pd.read_csv(join(input_dir, 'preprocessed_train.csv'))
test = pd.read_csv(join(input_dir, 'preprocessed_test.csv'))

train.head()
```

## Th·ª≠ nghi·ªám bi·∫øn TotalSF

### TotalSF = 1stFlrSF + 2ndFlrSF + TotalBsmtSF

```python
train['TotalSF'] = train['1stFlrSF'] + train['2ndFlrSF'] + train['TotalBsmtSF']
test['TotalSF'] = test['1stFlrSF'] + test['2ndFlrSF'] + test['TotalBsmtSF']

catboost(df=train, df_test=test)
xgboost(df=train, df_test=test)
```

### Th·ª≠ nghi·ªám lo·∫°i b·ªè th·ª≠ 3 bi·∫øn 1stFlrSF, 2ndFlrSF, TotalBsmtSF

```python
catboost(df=train.drop(['1stFlrSF', '2ndFlrSF', 'TotalBsmtSF'], axis=1), df_test=test.drop(['1stFlrSF', '2ndFlrSF', 'TotalBsmtSF'], axis=1))
xgboost(df=train.drop(['1stFlrSF', '2ndFlrSF', 'TotalBsmtSF'], axis=1), df_test=test.drop(['1stFlrSF', '2ndFlrSF', 'TotalBsmtSF'], axis=1))
```

### TotalSF = 1stFlrSF + TotalBsmtSF

```python
train['TotalSF'] = train['1stFlrSF'] + train['TotalBsmtSF']
test['TotalSF'] = test['1stFlrSF'] + test['TotalBsmtSF']

catboost(df=train, df_test=test)
xgboost(df=train, df_test=test)
```

### Th·ª≠ nghi·ªám lo·∫°i b·ªè th·ª≠ 2 bi·∫øn 1stFlrSF, TotalBsmtSF

```python
catboost(df=train.drop(['1stFlrSF', 'TotalBsmtSF'], axis=1), df_test=test.drop(['1stFlrSF', 'TotalBsmtSF'], axis=1))
xgboost(df=train.drop(['1stFlrSF', 'TotalBsmtSF'], axis=1), df_test=test.drop(['1stFlrSF', 'TotalBsmtSF'], axis=1))
```

Sau khi th·ª±c nghi·ªám xong, ta c√≥ th·ªÉ th·∫•y ƒë∆∞·ª£c `TotalSF = 1stFlrSF + TotalBsmtSF` l·∫°i c√≥ ƒë·ªô ch√≠nh x√°c cao h∆°n so v·ªõi `TotalSF = 1stFlrSF + 2ndFlrSF + TotalBsmtSF`

## N·ªôp file l√™n Kaggle

```python
catboost(
    df=train,
    df_test=test,
    author="Thien",
    path_to_log_csv=join('..', '..', 'log', 'TotalSF', 'experiment_logger.csv'),
    name_folder="TotalSF",
    save_log=True,
    save_model=True,
    save_submission=True
)

catboost(
    df=train.drop(['1stFlrSF', 'TotalBsmtSF'], axis=1), 
    df_test=test.drop(['1stFlrSF', 'TotalBsmtSF'], axis=1), 
    author="Thien", 
    path_to_log_csv=join('..', '..', 'log', 'TotalSF_drop', 'experiment_logger.csv'),
    name_folder="TotalSF_drop",
    save_log=True,
    save_model=True,
    save_submission=True
)
```
